### Example Starter Code


```python

# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
data = sns.load_dataset("titanic")

# Display First Few Rows
print(data.head())

# Print Dataset Information
print(data.info())

# dataset's shape, columns, and summary information
print(data.describe())

# Check for Missing Values
print(data.isnull().sum())





```

```python
#How does the survival rate vary by passenger class
sns.countplot(x='pclass', hue='survived', data=data)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
#Are survival rates different between males and females
sns.countplot(x='sex', hue='survived', data=data)
plt.title('Survival Rate Between Male and Female')
plt.show()
```

```python
#What is the distribution of ages among passengers
sns.histplot(data['age'], kde=True)
plt.title('Distribution of Ages')
plt.show()
```

```python
#How does the survival rate differ for passengers with family aboard
sns.countplot(x='sibsp', hue='survived', data=data)
plt.title('Survival Rate by Family size')
plt.show()
```
